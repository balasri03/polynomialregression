```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x=6*np.random.rand(100,1)-3
y=0.5*x**2+2+np.random.rand(100,1)
plt.scatter(x,y,color='g')
plt.xlabel(' x dataset')
plt.ylabel('y dataset')
print(x.shape,y.shape)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
from sklearn.metrics import r2_score
score=r2_score(y_test,regressor.predict(x_test))
print(score)
plt.plot(x_train,regressor.predict(x_train),color='r')
plt.scatter(x_train,y_train)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2,include_bias=True)
x_train_ploy=poly.fit_transform(x_train)
x_test_poly=poly.fit_transform(x_test)
from sklearn.metrics import r2_score
regressor_1=LinearRegression()
regressor_1.fit(x_train_ploy,y_train)
y_pred=regressor_1.predict(x_test_poly)
score=r2_score(y_test,y_pred)
print(score)
plt.scatter(x_train,regressor_1.predict(x_train_ploy),color='r')
plt.scatter(x_train,y_train)
```
